```
from keras.layers import Activation,Dense
import pandas as pd
import keras
import matplotlib.pyplot as plt
import numpy as np
network=keras.models.Sequential()
layer1=Dense(2,input_shape=(2,),activation='sigmoid')
layer2=Dense(1,activation='sigmoid')
network.add(layer1)
network.add(layer2)
optimiser=keras.optimizers.SGD(learning_rate=0.5,momentum=0.1)
network.compile(optimizer=optimiser,loss='mse',metrics=['mse'])
inp=pd.DataFrame([[0,0],[0,1],[1,0],[1,1]])
out=pd.DataFrame([0,1,1,0])
history=network.fit(inp,out,epochs=5000,batch_size=4)
```

```
print(layer1.get_weights())
print(network.predict(inp))
plt.plot(history.history['mse'])
plt.show()
```
