```
from keras.layers import Activation,Dense
import pandas as pd
import keras
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
from keras.utils import to_categorical
def decisionboundary(model):
    x_span=np.linspace(0,1,20)
    y_span=np.linspace(0,1,20)
    xx,yy=np.meshgrid(x_span, y_span)
    labels=model.predict(np.c_[xx.ravel(), yy.ravel()])
    z=labels.reshape(xx.shape)
    _,ax=plt.subplots()
    ax.contourf(xx,yy,z)
    plt.show()
```

```
network=keras.models.Sequential()
inp=np.array([[0,0],[0,1],[1,0],[1,1]])
out=np.array([0,1,1,0])
layer1=Dense(2,input_shape=(2,),activation='sigmoid')
layer2=Dense(1,activation='sigmoid')
network.add(layer1)
network.add(layer2)
optimiser=keras.optimizers.SGD(learning_rate=2,momentum=0.5)
network.compile(optimizer=optimiser,loss='mse',metrics=['mse'])
history=network.fit(inp,out,epochs=300,batch_size=4)
```

```
print(network.predict(inp))
decisionboundary(network)
```

```
andnetwork=keras.models.Sequential()
andinp=np.array([[0,0],[0,1],[1,0],[1,1]])
andout=np.array([0,0,0,1])
andlayer1=Dense(1,input_shape=(2,),activation='sigmoid')
andnetwork.add(andlayer1)
andoptimiser=keras.optimizers.SGD(learning_rate=1.5,momentum=0.4)
andnetwork.compile(optimizer=andoptimiser,loss='mse',metrics=['mse'])
andhistory=andnetwork.fit(andinp,andout,epochs=300,batch_size=4)
```
